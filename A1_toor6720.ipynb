## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('energy-price.csv')

# List of cities to compare
cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor', 'London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound', "St. Catharine's"]

# Initialize an empty dictionary to store mean prices for each city
mean_prices_dict = {}

# Iterate over the cities and calculate the mean prices of Regular Unleaded Gasoline for each city in 2023
for city in cities:
    mean_price = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df[city].notnull()) & (df['Date'].str.contains('2023'))][city].mean()
    mean_prices_dict[city] = mean_price

# Sort the cities by mean prices
sorted_cities = sorted(mean_prices_dict, key=lambda x: mean_prices_dict[x])

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(sorted_cities, [mean_prices_dict[city] for city in sorted_cities], color='skyblue')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.xlabel('City')
plt.ylabel('Mean Price (CAD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('energy-price.csv')

# Filter the data for Regular Unleaded Gasoline in Toronto West/Ouest for 2023
toronto_west_gas_2023 = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023'))]['Toronto West/Ouest']

# Sort the data
sorted_prices = np.sort(toronto_west_gas_2023)

# Calculate the ECDF
ecdf = np.arange(1, len(sorted_prices) + 1) / len(sorted_prices)

# Create a normalized ECDF plot
plt.figure(figsize=(10, 6))
plt.plot(sorted_prices, ecdf, marker='.', linestyle='none', color='skyblue')
plt.title('Normalized ECDF of Regular Unleaded Gasoline Prices in Toronto West/Ouest (2023)')
plt.xlabel('Price (CAD)')
plt.ylabel('ECDF')
plt.margins(0.02)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the plot
plt.show()
```

## Question 5

Plot a boxplot to display the distribution of Regular Unleaded Gasoline prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('energy-price.csv')

# List of cities to compare
cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor', 'London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound', "St. Catharine's"]

# Filter the data for Regular Unleaded Gasoline and from 2021 onwards
filtered_data = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2021'))]

# Create a boxplot
plt.figure(figsize=(12, 6))
plt.boxplot([filtered_data[city] for city in cities], labels=cities, vert=False, showfliers=False)
plt.title('Boxplot of Regular Unleaded Gasoline Prices from 2021 Onwards')
plt.xlabel('Price (CAD)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the plot
plt.show()
```
